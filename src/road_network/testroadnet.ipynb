```python
import create_graph as cg
import pandas as pd
import random
```

```python
graph = cg.initialise_graph('../../data/random_weight_edge.csv')
```

get random values for start and end point

```python
file_path =  '../../data/random_weight_edge.csv' 

df = pd.read_csv(file_path)
nodes = df["u"].to_list()
random_values = random.sample(nodes, 2)
print(random_values)
```

```python
# Example usage
G = cg.create_osmnx_compatible_graph(file_path)

route = cg.dijkstra(graph, random_values[0], random_values[1])

print(route)
# # Plot just the graph
fig, ax = cg.plot_graph_with_routes(G,route)
```
