```python
import json
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
with open("../test_data/test_output.json", "r") as file:
    complete_results_data = json.load(file)
```

```python
def calculate_cumulative_values(values:list)->list:
    cumulative_values = []
    running_sum = 0
    
    for value in values:
        running_sum += value
        cumulative_values.append(running_sum)
        
    return cumulative_values
```

```python
distances = []
energies = []
currents = []
times = []
climbs = []

#sort through the items in the results file:
for keys, dict in complete_results_data.items():

    #collect just section data
    if "path" in keys:
        for section_name, section_data in dict.items():
            # pprint(section_data)
            times.append(section_data["time"])
            distances.append(section_data["distance"])
            climbs.append(section_data['climb'])

cumulative_dist = calculate_cumulative_values(distances)
cumulative_time = calculate_cumulative_values(times)
cumulative_climb = calculate_cumulative_values(climbs)
```
